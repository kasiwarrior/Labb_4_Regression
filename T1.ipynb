##Uppgift 1.1 

Mean Absolute Error (MAE) är ett regressionsmått som visar hur stora felen i en modells prediktioner är i genomsnitt. Det beräknas genom att ta den absoluta skillnaden mellan de verkliga värdena och de predikterade värdena, och sedan räkna ut medelvärdet av dessa skillnader. Eftersom absoluta värden används spelar det ingen roll om felet är positivt eller negativt.
MAE är lätt att tolka eftersom resultatet är i samma enhet som målvariabeln. Till exempel betyder en MAE på 3 att modellens prediktioner i genomsnitt ligger 3 enheter från de verkliga värdena.
Jag skulle inte använda den här metoden om det är viktigt att straffa stora fel mer än små fel, eftersom MAE behandlar alla fel lika.

##Uppgift 1.2 

PolynomialFeatures är en metod i scikit-learn som används för att skapa nya features från befintliga features. Metoden gör detta genom att polynom, Tex genom att höja fetures till olika potenser eller genom att kombinera dem med varandra. Detta hjälper till att fånga icke-linjära samband i data som en enkel linjär modell kanske inte kan upptäcka.


##Uppgift 1.3

Om vi har inputen (x1,x2,x3,x4) och sätter interaction_only=True kommer PolynomialFeatures att generera nya features som endast är produkter av de ursprungliga features utan att inkludera deras potenser. Detta innebär att vi får alla möjliga kombinationer av de ursprungliga features multiplicerade med varandra, men inte deras kvadrater eller högre potenser.
